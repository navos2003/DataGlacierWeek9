# Data Cleaning, Mean and Median Imputation

```python
df.isnull().mean()
```

```python
missing_vars_numeric = [var for var in df.columns
                         if df[var].isnull().mean() > 0 and df[var].dtypes != '0']
```

```python
missing_vars_numeric
```

```python
#def distribution_plot(df, var):
#    fig = plt.figure()
#    ax = fig.add_subplot(111)
#    sns.kdeplot(df[var], color='b', shade=True, multiple='stack')
#    lines, labels = ax.get_legend_handles_labels()
#    ax.legend(lines, labels, lock='best')
#    plt.show()
def distribution_plot(df, var):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    sns.countplot(x=var, data=df)
    ax.set_title(var)
    plt.show()
```

```python
for var in missing_vars_numeric:
    distribution_plot(df,var)
```

```python
def impute_mean(df, var):
    df = df.copy()
    df[var + '_mean_imputed'] = df[var].fillna(df[var].mean())
    return df
```

```python
def impute_median(df, var):
    df = df.copy()
    df[var + '_median_imputed'] = df[var].fillna(df[var].median())
    return df
```

```python
df = impute_mean(df, 'cod_prov')
```

```python
df = impute_median(df, 'cod_prov')
```

```python
def plot_compare(df, var, var_inputed):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    sns.countplot(x=var, data=df)
    sns.countplot(x=var_inputed, data=df)
    ax.set_title(var)
    plt.show()
```

# Plot comparsion after imputation

```python
plot_compare(df, 'cod_prov', 'cod_prov_mean_imputed')
```
